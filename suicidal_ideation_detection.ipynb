## **Data Preprocessing**

### **Data Cleaning**

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Suicidal Ideation Detection/Data/ss_data.csv')
```

3. Defining cleaning functions.

```python
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

def remove_mentions(text):
    return re.sub(r'@\w+', '', text)

def remove_hashtags(text):
    return re.sub(r'#\w+', '', text)

def remove_digits(text):
    return re.sub('\d+', '', text)

def remove_non_english(text):
    return re.sub('[^\x00-\x7F]+', ' ', text)

def remove_punctuation(text):
    return text.translate(str.maketrans(string.punctuation, ' '*len(string.punctuation)))

def translate_emoji(text):
    return demoji.replace_with_desc(text, sep=" ")

def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    return [word for word in text if word not in stop_words]

def tokenize(text):
    return word_tokenize(text)

def lemmatize(text):
    lemmatizer = WordNetLemmatizer()
    lemmatized = [lemmatizer.lemmatize(token) for token in text]
    return lemmatized
```

4. Apply cleaning functions to the `text` column.

```python
# Converting text to lowercase.
df['text'] = df['text'].apply(lambda x: x.lower())

# Remove urls from the text.
df['text'] = df['text'].apply(remove_urls)

# Remove any mentions '@' from the text.
df['text'] = df['text'].apply(remove_mentions)

# Remove any hashtags '#' from the text.
df['text'] = df['text'].apply(remove_hashtags)

# Remove any numbers from the text.
df['text'] = df['text'].apply(remove_digits)

# Remove non-english language such as other languages characters.
df['text'] = df['text'].apply(remove_non_english)

# Remove punctuation such as commas and exclamation marks etc.
df['text'] = df['text'].apply(remove_punctuation)

# Convert emoji into text description.
df['text'] = df['text'].apply(translate_emoji)

# Remove stopwords such as 'a' and 'the'.
df['text'] = df['text'].apply(remove_stopwords)

# Tokenize sentences into chunks.
df['text'] = df['text'].apply(tokenize)

# Lemmatize word to its root form.
df['text'] = df['text'].apply(lemmatize)

# Join the list of words back into a sentence.
df['text'] = df['text'].apply(' '.join)

# Replace any NaN values.
df['text'] = df['text'].replace('', np.nan)
df = df.dropna(subset=['text'])

df.to_csv('/content/drive/MyDrive/Colab Notebooks/Suicidal Ideation Detection/Data/ss_cleaned.csv', index=False)
```

2. Load the `sucide_sentiment_data_cleaned.csv` from `/content/drive/MyDrive/Colab Notebooks/suicide_sentiment_data_cleaned.csv`

```python
# Load pre-trained model and tokenizer
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = BertModel.from_pretrained('bert-base-uncased').to(device)  # move model to GPU
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Read data
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Suicidal Ideation Detection/Data/ss_cleaned.csv')

# Get list of texts from DataFrame
texts = df['text'].tolist()

# Define batch size
batch_size = 50

# Initialize list to hold embeddings
embeddings = []

# Process texts in batches
for i in range(0, len(texts), batch_size):
    batch_texts = texts[i:i+batch_size]

    # Encode batch
    inputs = tokenizer(batch_texts, truncation=True, max_length=512, padding='max_length', return_tensors="pt").to(device)

    # Get embeddings for batch
    with torch.no_grad():
        outputs = model(**inputs)
    batch_embeddings = outputs.last_hidden_state[:,0,:].cpu().numpy()

    # Add batch embeddings to list
    embeddings.extend(batch_embeddings)

# Convert list of embeddings to pandas DataFrame, then to cuDF DataFrame
embeddings_df = cudf.DataFrame(pd.DataFrame(embeddings))

# Convert the 'class' column to a cuDF DataFrame
class_df = cudf.Series(df['class']).to_frame()

# Concatenate along the column axis
final_df = cudf.concat([class_df, embeddings_df.reset_index(drop=True)], axis=1)

# Save the final DataFrame into a pickle file
final_df.to_pandas().to_pickle('/content/drive/MyDrive/Colab Notebooks/Suicidal Ideation Detection/Data/ss_bert.pkl')
```

### **Random Forest**


```python
from cuml.ensemble import RandomForestClassifier
from cuml.model_selection import train_test_split
from cuml.metrics import accuracy_score
from sklearn.model_selection import KFold
from sklearn.metrics import precision_score, recall_score, f1_score

# Load the pickle file into a pandas DataFrame
df_pd = pd.read_pickle('/content/drive/MyDrive/Colab Notebooks/Suicidal Ideation Detection/Data/ss_bert.pkl')

# Convert the pandas DataFrame to a cuDF DataFrame
df_cudf = cudf.DataFrame.from_pandas(df_pd)

X = df_cudf.drop(columns='class')
y = df_cudf['class']

# Split data into training set and test set
X, X_test, y, y_test = train_test_split(X, y, test_size=0.4, stratify=y)

# Create a RandomForestClassifier model
rf = RandomForestClassifier(n_estimators=100, max_depth=10)

# Setup KFold cross-validation
kf = KFold(n_splits=6, shuffle=True)

# Perform cross-validation
scores = []
for train_index, val_index in kf.split(X):
    X_train, X_val = X.iloc[train_index], X.iloc[val_index]
    y_train, y_val = y.iloc[train_index], y.iloc[val_index]

    # Fit model on training data
    rf.fit(X_train, y_train)

    # Score model on validation data
    val_pred = rf.predict(X_val)
    score = accuracy_score(y_val, val_pred)
    scores.append(score)

# Print mean cross-validation score
print(f"Mean cross-validation score: {np.mean(scores)}")

# Fit the model on the entire training data and evaluate on the test set
rf.fit(X, y)
test_pred = rf.predict(X_test)
test_score = accuracy_score(y_test, test_pred)

print(f"Accuracy: {test_score}")

# Convert cuDF Series to pandas Series for use with sklearn
test_pred_pd = test_pred.to_pandas()
y_test_pd = y_test.to_pandas()

# Calculate and print precision, recall, and f1 score
precision = precision_score(y_test_pd, test_pred_pd, average='weighted')
recall = recall_score(y_test_pd, test_pred_pd, average='weighted')
f1 = f1_score(y_test_pd, test_pred_pd, average='weighted')

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")

# Save model
dump(rf, '/content/drive/MyDrive/Colab Notebooks/Suicidal Ideation Detection/Models/rf_model.joblib')

```

### **XGBoost**

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd

# Load the pickle file into a pandas DataFrame
df_pd = pd.read_pickle('/content/drive/MyDrive/Colab Notebooks/Suicidal Ideation Detection/Data/ss_bert.pkl')

X = df_pd.drop(columns='class')
y = df_pd['class']

# Split data into training set and test set
X, X_test, y, y_test = train_test_split(X, y, test_size=0.4, stratify=y)

# Specify parameters
params = {
    'max_depth': 6,
    'learning_rate': 0.1,
    'objective': 'binary:logistic',
    'eval_metric': 'logloss',
    'tree_method': 'gpu_hist'  # use GPU
}

# Setup KFold cross-validation
kf = KFold(n_splits=6, shuffle=True)

# Perform cross-validation
scores = []
for train_index, val_index in kf.split(X):
    X_train, X_val = X.iloc[train_index], X.iloc[val_index]
    y_train, y_val = y.iloc[train_index], y.iloc[val_index]

    # Convert data into DMatrix which is a requirement for XGBoost
    dtrain = xgb.DMatrix(X_train, label=y_train)
    dval = xgb.DMatrix(X_val, label=y_val)

    # Train model on training data
    model = xgb.train(params, dtrain,num_boost_round=100)

    # Score model on validation data
    val_preds = model.predict(dval)
    val_preds_binary = [1 if p > 0.5 else 0 for p in val_preds]  # XGBoost outputs probabilities by default
    score = accuracy_score(y_val, val_preds_binary)
    scores.append(score)

# Print mean cross-validation score
print(f"Mean cross-validation score: {np.mean(scores)}")

# Convert test data into DMatrix
dtest = xgb.DMatrix(X_test, label=y_test)

# Fit the model on the entire training data and evaluate on the test set
dtrain_full = xgb.DMatrix(X, label=y)
model = xgb.train(params, dtrain_full)
test_preds = model.predict(dtest)
test_preds_binary = [1 if p > 0.5 else 0 for p in test_preds]

test_score = accuracy_score(y_test, test_preds_binary)

print(f"Test set score: {test_score}")

# Calculate and print precision, recall, and f1 score
precision = precision_score(y_test, test_preds_binary, average='weighted')
recall = recall_score(y_test, test_preds_binary, average='weighted')
f1 = f1_score(y_test, test_preds_binary, average='weighted')

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")

```
